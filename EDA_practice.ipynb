```python
#Bar Plot

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Values': [10, 10, 15]
})

# Set style and color palette
sns.set_style("darkgrid")  # Options: white, dark, whitegrid, darkgrid, ticks
sns.set_palette("colorblind")   # Try: deep, muted, bright, pastel, dark, colorblind

# Create the plot
plt.figure(figsize=(8, 5))
barplot = sns.barplot(x='Category', y='Values', data=data)

# Add value labels on top of bars
for p in barplot.patches:
    barplot.annotate(f'{p.get_height()}', 
                     (p.get_x() + p.get_width() / 2, p.get_height()), 
                     ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

# Customize title and labels
plt.title("Bar Plot of Categories", fontsize=14, fontweight='bold', color='navy')
plt.xlabel("Category", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Optional: Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optional: Remove top and right borders
sns.despine()

# Show the plot
plt.show()
```

```python
# bar plot (works on continuous dataset )

import pandas as pd

data = pd.DataFrame({
    "Class":['A','B','C','D','E'],
    "Students" :[90,46,35,67,55]
})

sns.barplot(x='Class', y='Students', data=data) 
plt.title("Bar Plot") 
plt.xlabel("Class")
plt.ylabel("Students")
plt.show() 
```

```python
# count plot-A count plot is a type of bar plot that shows the count of observations in each category of a categorical variable.
sns.countplot(x='Category', data=data) 
plt.title("Count Plot") 
plt.show() 
```

```python
# count plot-A count plot is a type of bar plot that shows the count of observations in each category of a categorical variable.
# sns.countplot(x='Category', data=data) 
# plt.title("Count Plot") 
# plt.show() 


import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C', 'B', 'B']
})

# Plot
plt.figure(figsize=(7, 5))
sns.set_style("whitegrid")
sns.countplot(x='Category', data=data, palette='pastel')

plt.title("Count Plot of Categories", fontsize=14, fontweight='bold')
plt.xlabel("Category")
plt.ylabel("Count")

plt.show()
```

```python
# Loading dataset

df = sns.load_dataset('iris') 
df
```

```python
# Pairplot 

sns.pairplot(df, hue='species') 
plt.show() 
```

```python
# Drop Categorical column

df_numeric = df.drop(columns='species')     # ✅ recommended
# or
df_numeric = df.drop('species', axis=1)     # ✅ alternate form
correlation_matrix = df_numeric.corr()
correlation_matrix
```

```python
# Heatmap  (To show intercorrelation between numerical variables)

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm') 
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pie chart  

labels = ['A', 'B', 'C'] 
sizes = [30, 50, 20] 

plt.pie(sizes, labels=labels, autopct='%1.1f%%',startangle =0)    
plt.title("Pie Chart") 
plt.show() 
```

```python
# Swarm plot

sns.swarmplot(x='species', y='sepal_length', data=df) 
plt.title("Swarm Plot") 
plt.show() 
```

```python
# Strip plot

sns.stripplot(x='species', y='sepal_length', data=df, jitter=True) 
plt.title("Strip Plot") 
plt.show()
```

```python
# Joint plot  (Scatter+Histogram)

sns.jointplot(x='sepal_length', y='petal_length', data=df, kind='scatter') 
plt.show()
```
